# Fall 2025 DS UCSB Instagram Analysis Dashboard

This notebook cleans and visualizes Instagram engagement data for DS UCSB from the months of August to December 2025. Data was obtained from Meta Business Suite.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from src.cleaning import strip_col_names, drop_cols, rename_cols, modify_cats, sort_published, sep_time, sort_inter

BASE_DIR = Path.cwd()
csv_path = BASE_DIR / "data" / "f25_instagram_data.csv"
save_path = "C:/Users/makalamesina-forester/Downloads/"

df = pd.read_csv(csv_path)
df.head()
```

```python
df = pd.read_csv(csv_path)
df.head()
```

```python
# Strip column names
df = strip_col_names(df)
df.head()
```

```python
# Drop irrelevant columns
df = drop_cols(df)
df.head()
```

```python
# Rename columns
df = rename_cols(df)
df.head()
```

```python
# Change certain data types and modify category names
df = modify_cats(df)
df.head()
```

```python
# Sort by Published date and time
df = sort_published(df)
df.head()
```

```python
# Create two new columns: Date and Time
df = sep_time(df)
df.head()
```

```python
# Rearrange column order
first = ["Date", "Time", "Post", "Category"]
rest = df.loc[:, "Duration (sec)":"Saves"].columns
df = df[list(first) + list(rest)]
df.head()
```

```python
# Create Interaction variable
df = df.assign(Interactions = df.iloc[:,8:12].sum(axis=1))
df.head()
```

```python
# Create copy of df, sort by total Interactions
df2 = sort_inter(df)
df2.head()
```

```python
# Plot top posts by Interaction
top_int = df2.iloc[0:5,:]
top_int_plot = sns.barplot(x="Post",
                           y="Interactions",
                           data=top_int)

top_int_plot.set_xticklabels(
    labels=["Membership Reel", 
            "Career Fair General", 
            "Fall Kickoff", 
            "Officer Introduction",
            "Intern Introductions"], rotation=30)


plt.savefig('top_ints1.png', bbox_inches = 'tight')
```

```python
# Plot Interactions by Views
int_by_views = sns.scatterplot(x="Views",
                               y="Interactions",
                               data=df2)

plt.savefig('int_by_views.png', bbox_inches = 'tight')
```
